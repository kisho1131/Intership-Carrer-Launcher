```python
def average_price(N):
    return (df['Average Price'].tail(N).sum())/N
print("Average prices for last N days are:")
print("Last 1 week",average_price(5))
print("Last 2 weeks",average_price(10))
print("Last 1 month",average_price(20))
print("Last 3 months",average_price(60))
print("Last 6 months",average_price(120))
print("Last 1 year",average_price(240))
def prof_loss(N):
    difference = (df['Close Price'].tail(N).iloc[N-1] - df['Close Price'].tail(N).iloc[0])
    if difference < 0 :
        loss = -(difference)
        loss_percen = (loss/df['Close Price'].tail(N).iloc[N-1])*100
        return loss_percen
    if difference > 0 :
        profit = difference
        profit_percen = (profit/df['Close Price'].tail(N).iloc[N-1])*100
        return profit_percen
print("Profit-Loss percentage for last N days:")
print("Last 1 week",prof_loss(5))
print("Last 2 weeks",prof_loss(10))
print("Last 1 month",prof_loss(20))
print("Last 3 months",prof_loss(60))
print("Last 6 months",prof_loss(120))
print("Last 1 year",prof_loss(240))
```
